1. Provide at least two differences between the NumPy and Pandas packages.

1. **NumPy** (Numerical Python) provides multidimensional arrays and efficient numerical operations. **Pandas** (Panel Data) is built on top of NumPy and adds user-friendly features such as named rows and columns, missing data management, and the ability to read/write files in various formats. 
2. NumPy's primary data structure is the `Array`, while Pandas uses `DataFrame` and `Series`. Pandas DataFrames are essentially NumPy Arrays with additional capabilities like grouping, merging, and joining.

2. How is "dot notation" is used in Python?

Dot notation uses the period (`.`) to chain an object to its attributes or methods. The syntax is `object.attribute` or `object.method()`. It is used for three main purposes in Python: (1) accessing methods or attributes of an object (e.g., `s.upper()`), (2) accessing contents of a module or package (e.g., `math.sqrt(16)`), and (3) accessing columns or methods of a DataFrame (e.g., `df.population`).

3. What do the `dir()` and `type()` functions do in Python?

`type()` returns the type (class) of an object. For example, `type(2)` returns `int`, `type('dog')` returns `str`, and `type(df)` returns `pandas.core.frame.DataFrame`. `dir()` lists all the attributes and methods available on an object. For example, `dir(s)` on a string shows all string methods like `upper`, `lower`, etc. Called without arguments, `dir()` shows all objects defined in the current session.

4. What does "namespace" mean in Python and why is it important to keep different packages in their own namespace?

A namespace is a collection of currently defined names along with information about the object each name references. It is important to keep different packages in their own namespaces to prevent name collisions. For example, both `math` and `numpy` have a `sqrt` function — by keeping them in separate namespaces (`math.sqrt` vs `np.sqrt`), we can use both without conflict. This is why `from module import *` is discouraged, as it dumps all names into the main namespace and can silently overwrite existing functions.

5. What does it mean when a programming language is said to be "zero-indexed"? Is Python zero-indexed?

A zero-indexed language means that the first element of a sequence is at position 0, the second element is at position 1, and so on. Yes, Python is zero-indexed. For example, given `s = 'hello'`, `s[0]` returns `'h'` (the first character) and `s[1]` returns `'e'` (the second character).

6. Once a function is defined, you can then run it. When running a function (not defining a function), why can you sometimes skip the argument name and sometimes you cannot?

You can skip the argument name when passing arguments in the correct positional order defined by the function signature. If you pass arguments out of order, you must use the argument names (keyword arguments). Also, once you start naming arguments, all subsequent arguments must also be named. For example, with `round(number, ndigits)`, you can call `round(10.867, 2)` positionally, or `round(ndigits=2, number=10.867)` using keyword arguments in any order.
