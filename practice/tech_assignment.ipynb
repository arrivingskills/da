# Jupyter Notebook Intro

We will make extensive use of Jupyter notebooks in this course. The goal here is to give you some practice running a notebook in VS Code. Most of the code is provided, but there are places where you need to write brief snippets of code.

### Notebook Structure

A notebook is made up cells. We will use two types of cells in this course.
- Markdown cells
  - Contain Markdown code
  - Markdown is a language for writing nicely formatted text
  - This is a Markdown cell
- Code cells
  - Contain Python code that can be run
  - The output of the code appears below the cell

### Using a Jupyter Notebook
- How to move between cells to make them active?
  - Option 1: use the up and down arrow keys on the keyboard (recommended)
  - Option 2: click on a cell
- How to run the active cell?
  - Option 1: hold the Shift key and hit Enter (recommended)
  - Option 2: click the `Run` button (right facing triangle) on the left of the cell
- How to edit the active cell?
  - Option 1: hit the Enter key
  - Option 2: click inside a code cell
  - Option 3: double-click inside a Markdown cell
