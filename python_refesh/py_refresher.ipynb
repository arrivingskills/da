### Modules, Packages and Libraries

Earlier in this notebook we mentioned that Python has a small core language with additional functionality available from outside. In this section we discuss how to access that outside functionality.

First, some vocabulary:
- A "module" is a single file containing Python code
- A "package" or "library" is a group of files containing Python code
- Modules, packages and libraries:
  - Can define functions, classes and variables
  - Usually include runnable code
  - Typically created by other people

#### Importing

There are a few ways to import (same for modules, packages and libraries)
- `import ____`
  - `import module_name` (imports one module)
  - `import module1, module2, module3` (imports multiple modules)
- `import ____ as ____`
  - `import module_name as mod` (imports one module and assigns it an alias; usually used to give it a shorter name)
- `from ____ import ____`
  - `from module_name import object_name` (imports a specific object from a module)
  - `from module_name import object1, object2, object3` (imports multiple objects from a module)
- `from ____ import ____ as ____`
  - `from module_name import object_name as obj` (imports a specific object from a module and assigns the object an alias)
- `from ____ import *`
  - `from module_name import *` (imports all the module's objects into the main namespace)
  - (almost) never do this; we'll show why in some examples below
  
Again, all the above cases work the same for modules, packages and libraries.

#### Accessing the contents

Each module, package or library you import has its own "namespace" that contains all its objects. 

> A namespace is a collection of currently defined names along with information about the object that each name references.

Let's check out another package. If you remember from near the beginning of this notebook, we discussed NumPy as one of the key data analytics packages in Python. In some ways it is a competitor to the `math` library. Let's see why.

````python
import numpy as np
````

The Python community almost always uses the `np` alias for `numpy`. We will also use this convention.

````python
np.sqrt(16)
````

````python
math.sqrt(16)
````

Notice that both NumPy and `math` have square root functions. We can differentiate the two because the NumPy version of square root lives in the `np` namespace and the `math` version lives in the `math` namespace. 

Why bother keeping them separate? Square root is square root, right? Kinda. One reason we really like NumPy is that it allows us to perform mathematical operations on arrays. We will discuss array objects later, but for now consider it a group of numbers. Check out the following cell.

````python
np.sqrt([16, 25])
````

We pass in `16` and `25` and get back `4.` and `5.`. Pretty cool, right? Two numbers go in, two square roots come out. But notice what happens if we pass two number to the `math` version of `sqrt`.

````python
math.sqrt([16, 25])
````

The `math` version can only work on one number at a time. This is why we keep the namespaces separate for the different packages we import.

Remember I said that we should never use the following syntax?
```python
from module_name import *
```
This command imports all the objects from the package into the main namespace, without the benefit of keeping the contents of different packages separate. Can you see the practical problem with running the following code in the same notebook?
```python
from math import *
from numpy import *
```
Both of these packages have a `sqrt` function. Since we import NumPy second, it will overwrite the `sqrt` from the `math` package. When working on real world projects, you might import many packages and it's reasonable to assume that the authors of those various packages might use the same object names and thus importing them all into the same namespace will result in conflicts.

My recommendation is to __never__ use the `import *` syntax.

### Flow Control

Python has...
- `if` statements
- `for` loops 
- `while` loops 

#### Code blocks

Syntax   
- A colon (`:`) at the end of a line starts a code block
- The next line (i.e., the first line of the code block) must be indented
  - All subsequent lines in the code block must have the __same__ indentation level
  - The code block ends on the first line that is not indented

Indentation   
- Spaces or tabs are fine ([although this can be a point of contention](https://youtu.be/oRva7UxGQDw))
- The Python convention is 4 spaces (very strong recommendation)
- Cannot mix and match in one code block

#### `if` ... `elif` ... `else`

This type of flow must start with `if` followed by some condition. If the condition evaluates to `True`, then the subsequent indented code block is run. If the condition evaluates to `False`, then Python looks for an `elif` or `else` statement and runs those. It's important to know that `elif` and `else` are not required.

````python
if 7 > 3:
    print('if print')
else:
    print('else print')
````

In the example above, since `7` is greater than `3`, the code block under the `if` statement was triggered

````python
if 7 < 3:
    print('if print')
else:
    print('else print')
````

In the example above, the `if` statement evaluates to `False` so the code block under the `else` is run

The cell below shows that variables can be passed to `if` statements

````python
x = 2

if x == 4:
    print('it matches')
````

Why did the cell above not produce any output???

The example below demonstrates the use of `elif`. An `elif` must be followed by a conditional, just like `if`. The `else` at the end is the catch-all if none of the preceding `if` or `elif` statements are triggered.

````python
if x == 0:
    print("x equals 0")
elif x == 1:
    print("x equals 1")
elif x == 2:
    print("x equals 2") 
elif x == 3:
    print("x equals 3")
else:
    print("x greater than 3")
````

#### `for` loop

A `for` loop iterates over a sequence. For each element in the sequence it runs the code in the code block below the `for` statement.

In the following example, each element in the list is named `ingredient` one-by-one and then is available to use in the code block.

````python
ingredients = ['eggs', 'chicken', 'salad', 'doughnut']

for ingredient in ingredients:
    print(ingredient + ' with bacon')
````

So what's happening here?
- We first define a list of food items called `ingredients`
- The `for` loop runs four times, once for each element in the list
- The first time through, `ingredient` is `eggs`, so the code block prints `eggs with bacon`
- The second time through, `ingredient` is `chicken`, so the code block prints `chicken with bacon`
- The `for` loop stops when it gets to the end of the list

Before you ask, the answer is: "[yes, you can put bacon on a doughnut](https://sallysbakingaddiction.com/wp-content/uploads/2020/03/homemade-maple-bacon-doughnuts.jpg)"

We can nest the various flow controls in one another. However, since Python is sensitive to the indentation, the coder needs to keep track of the indentation levels. Below is a nested example.

````python
ingredients
````

````python
for ingredient in ingredients:
    if ingredient == 'salad':
        print(ingredient.upper())
    else:
        print(ingredient)
````

In the example above, the print statements are indented 8 spaces since they are code blocks under the `if` and `else` statements, which themselves are indented 4 spaces since they're code blocks under the `for` statement.

#### Indexing a DataFrame

Above we showed how to get a particular column. What if we want a particular value from the DataFrame?

````python
az.loc['Coconino','h_units']
````

Look closely at the code above. The `loc` attribute of a DataFrame, followed by square brackets provides access to any value in the DataFrame. It is organized as `[row_values, column_values]`.

This notation is quite flexible and allows us to nest identifiers as in the cell below.

````python
az.loc[['Coconino','Yavapai'],['vacant','occupied']]
````

In the cell above, square brackets are nested in square brackets. In this example we request two specific rows and two specific columns resulting in a 2x2 DataFrame being returned.

If we leave either the row or column area blank, Pandas interprets that to mean we want everything. In the following example we leave the columns area blank so we get all the column values for `Coconino` county.

````python
az.loc['Coconino',]
````

We can use operators with the `.loc` syntax to extract rows of the DataFrame based on a condition. In the example below, we want to extract all the counties (i.e., rows) with population greater than 200,000.

````python
az.loc[az['population']>200000,]
````

Let's pull this cell apart:
- `az`: a DataFrame
- `.loc[xxx,yyy]`: allows us to extract rows and columns from a DataFrame with surgical precision 
  - `xxx`: the first term in the square brackets defines the rows we want
  - `yyy`: the second term in the square brackets defines the columns we want
- `az['population']>200000`: in English this says, "look in the `population` column, and give me all the rows with a value greater than `200000`"; this is sometimes called a "mask" since it masks out all the rows that don't meet the condition
- The column part (i.e., the `yyy`) is blank, which means "give me all the columns"

The mask can in principle be any condition. 

Not to freak you out, but it's useful to see how powerful these `.loc` requests can be. In words, what is the following command asking?

````python
az.loc[(az['population']>150000) & (az['vacant']<30000), ['population','vacant']]
````
